```python
import pandas as pd
```

```python
pd.set_option('display.max_columns', 50)

data = pd.read_csv('/content/cumulative.csv')
data.head()
```

```python
unrelated_features = ['rowid', 'kepid', 'kepoi_name', 
                       'kepler_name', 'koi_disposition', 'koi_insol', 
                       'koi_insol_err1', 'koi_insol_err2', 'koi_slogg', 
                       'koi_slogg_err1', 'koi_slogg_err2', 'koi_fpflag_ec',
                      'koi_teq', 'koi_teq_err1', 'koi_teq_err2', 'koi_tce_delivname']
data = data.drop(unrelated_features, axis=1)
```

```python
data = data.dropna()
```

```python
from sklearn import preprocessing
```

```python
le = preprocessing.LabelEncoder()
```

```python
data['koi_pdisposition'] = le.fit_transform(data['koi_pdisposition'])
```

```python
data_features = data.drop(['koi_pdisposition'], axis=1)

data_features.head()
```

```python
data_target = data[['koi_pdisposition']]
```

```python
from sklearn.model_selection import train_test_split
import torch 
from torch.utils.data import DataLoader, TensorDataset
import numpy as np
```

```python
x_train, x_test, y_train, y_test = train_test_split(data_features, 
                                                    data_target, 
                                                    test_size=0.2, 
                                                    random_state=0)
```

```python
x_train = torch.from_numpy(x_train.values).float()
x_test = torch.from_numpy(x_test.values).float()

y_train = torch.from_numpy(y_train.values).view(1,-1)[0]
y_test = torch.from_numpy(y_test.values).view(1,-1)[0]
```

```python
import torch.nn as nn
import torch.nn.functional as F
```

```python
x_train.shape, y_train.shape
```

```python
input_size = 33
output_size = 2
hid1_size = 25
hid2_size = 10
```

```python
class Net(nn.Module):
  def __init__(self):
    super(Net, self).__init__()
    self.fc1 = nn.Linear(input_size, hid1_size)
    self.fc2 = nn.Linear(hid1_size, hid2_size)
    self.fc3 = nn.Linear(hid2_size, output_size)
  
  def forward(self, x):
    x = F.sigmoid(self.fc1(x))
    x = F.sigmoid(self.fc2(x))
    x = self.fc3(x)

    return F.log_softmax(x, dim=-1)
```

```python
model = Net()
```

```python
import torch.optim as optim

optimizer = optim.Adam(model.parameters())

loss_fn = nn.NLLLoss()
```

```python
epochs = 1001
```

```python
for epoch in range(1, epochs):
  optimizer.zero_grad()

  output_train = model(x_train)

  loss = loss_fn(output_train, y_train)
  loss.backward()

  optimizer.step()

  # testing data
  output_test = model(x_test)
  loss_test = loss_fn(output_test, y_test)

  _, pred = output_test.data.max(1)

  # accuracy
  accuracy = torch.sum(pred == y_test.data).item() / y_test.size(0)

  if epoch % 50 == 0:
    print (f'epoch: {epoch} || train loss: {loss} || test loss: {loss_test} || accuracy: {accuracy}')
```
